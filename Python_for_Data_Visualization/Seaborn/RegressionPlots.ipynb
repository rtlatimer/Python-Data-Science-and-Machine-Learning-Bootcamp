# Regression Plots

Seaborn has many built-in capabilities for regression plots, however we won't really discuss regression until the machine learning section of the course, so we will only cover the **lmplot()** function for now.

**lmplot** allows you to display linear models, but it also conveniently allows you to split up those plots based off of features, as well as coloring the hue based off of features.

Let's explore how this works:

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot()

```python
# Create an lmplot(). Specify features x & y axes
sns.lmplot(x='total_bill',y='tip',data=tips)
```

It basically creates a scatter plot with a regression line on top of it.

```python
# We can define 'hue' to view the split between categorical data
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

Male and Female have basically have the same linear fit.

```python
# Can change our color palette
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='magma')
```

### Working with Markers

lmplot kwargs get passed through to **regplot** which is a more general form of lmplot(). regplot has a scatter_kws parameter that gets passed to plt.scatter. So you want to set the s parameter in that dictionary, which corresponds (a bit confusingly) to the squared markersize. In other words you end up passing a dictionary with the base matplotlib arguments, in this case, s for size of a scatter plot. In general, you probably won't remember this off the top of your head, but instead reference the documentation.

```python
# We can also change the markers 
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='magma',
           markers=['o','v'])
```

```python
# We can also adjust the size of the markers by making a call to a matplotlib argument
# Set 'scatter_kws' and define 's' (scatter)
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='magma',
           markers=['o','v'],scatter_kws={'s':100})
```

## Using a Grid

We can add more variable separation through columns and rows with the use of a grid. Just indicate this with the col or row arguments:

```python
# Instead of defining 'hue' which plotted the data for both sexes on the same graph,
# we can define 'col' to create a second graph, so the data isn't overlapping
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
# Can also split further by defining a 'row'
sns.lmplot(x="total_bill", y="tip", row="time", col="sex",data=tips)
```

Now we see the difference between male and female during both lunch and dinner.

```python
# Here we combine 'hue' and 'col'
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='magma')
```

## Aspect and Size

Seaborn figures can have their size and aspect ratio adjusted with the **size** and **aspect** parameters:

```python
# The plot above is very small and tough to read, so we can adjust 'aspect' and 'size'
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='magma',
          aspect=0.6,size=8)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='magma',
          aspect=0.8,size=6)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='magma',
          aspect=0.2,size=12)
```
