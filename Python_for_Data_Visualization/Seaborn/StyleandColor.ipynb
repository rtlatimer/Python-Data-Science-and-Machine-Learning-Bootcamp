# Style and Color

We've shown a few times how to control figure aesthetics in seaborn, but let's now go over it formally:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

## Styles

You can set particular styles.
There are 5 preset style themes:

#### 1. Darkgrid
This is the default style, so we don't actually need to specify it.

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

#### 2. Whitegrid

```python
# Use sns.set_style
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

#### 3. Dark

```python
sns.set_style('dark')
sns.countplot(x='sex',data=tips,palette='deep')
```

#### 4. White

```python
sns.set_style('white')
sns.countplot(x='sex',data=tips,palette='deep')
```

#### 5. Ticks

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

## Spine Removal

```python
# despine() allows you to remove spines.
# It only removes the top and right by default
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
# To remove the left, set 'left' equal to True
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```

## Size and Aspect

You can use matplotlib's **plt.figure(figsize=(width,height) ** to change the size of most seaborn plots.

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. For example:

```python
# Adjust figsize from matplotlib in conjunction with Seaborn
# Non Grid Plot
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',size=2,aspect=4,data=tips)
```

## Scale and Context

The set_context() allows you to override default parameters.
There are 4 preset contexts:

#### 1. Paper

```python
sns.set_context('paper',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```

#### 2. Notebook

```python
sns.set_context('notebook',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```

#### 3. Talk

```python
sns.set_context('talk',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```

#### 4. Poster

```python
sns.set_context('poster',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```

Check out the documentation page for more info on these topics:
