## FiveThirtyEight Style Plots

```python
# import pandas
import pandas as pd
```

```python
# load our dataset and view head
# This dataset describes the degrees earned by women from 1970 to 2011
direct_link = 'http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv'
women_majors = pd.read_csv(direct_link)

print(women_majors.info())
women_majors.head()
```

```python
women_majors.tail()
```

From a brief glance, it looks like engineering and computer science have seen a steady increase, but still a ways to go.

```python
# Return majors with less than 20% from 1970
under_20 = women_majors.loc[0, women_majors.loc[0] < 20]
under_20
```

```python
# Let's draw a simple graph of the trend in majors over time
%matplotlib inline
under_20_graph = women_majors.plot(x = 'Year', y = under_20.index, figsize = (12,8))
print('Type:', type(under_20_graph))
```

```python
# Let's import preset styles
import matplotlib.style as style
style.available
```

```python
# Set the 'fivethirtyeight' style and redraw the plot
style.use('fivethirtyeight')
women_majors.plot(x = 'Year', y = under_20.index, figsize = (12,8))
```

Well that's a bit more vibrant!
